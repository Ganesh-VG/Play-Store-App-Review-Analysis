## 3. ***Data Wrangling***

### Data Wrangling Code

````python
storedata_df = storedata_df.drop_duplicates(subset=['App', 'Rating', 'Size', 'Installs', 'Type',
       'Price', 'Genres', 'Current Ver',
       'Android Ver'])
````

````python
# Function to convert value in size column to float
def convert_to_float_for_size(value):
    if 'M' in value:
        return float(value.replace('M', ''))
    return None

# Apply the conversion function to the 'amount' column
storedata_df['Size'] = storedata_df['Size'].apply(convert_to_float_for_size)
````

````python
# Rename the column from 'Size' to 'Size(MB)'
storedata_df.rename(columns={'Size': 'Size(MB)'}, inplace=True)
````

````python
# Define the function to convert the reviews to integers
def to_integer(value):
    # Remove 'M' or 'B' and convert to integer
    value = str(value)
    if 'M' in value:
      return int(value.replace('.0M', '')) *1000000
    elif 'B' in value:
      return int(value.replace('.0B', '')) *1000000
    else:
      return int(value)

# Apply the conversion function to the 'amount' column
storedata_df["Reviews"] = storedata_df["Reviews"].apply(to_integer)
````

````python
# Function to convert value in price column to float
def convert_to_float_for_price(value):
    if '$' in value:
      return float(value.replace('$', ''))
    elif value == "Everyone":
      return None
    else:
      return float(value)

# Apply the conversion function to the 'amount' column
storedata_df["Price"] = storedata_df["Price"].apply(convert_to_float_for_price)
````

````python
# Rename the column from 'Price' to 'Price($)'
storedata_df.rename(columns={'Price': 'Price($)'}, inplace=True)
````

````python
# Remove non-numeric characters (excluding '+') and convert to integer
storedata_df['Installs'] = storedata_df['Installs'].str.replace('[,]', '', regex=True)
storedata_df = storedata_df[storedata_df['Installs'] != 'Free']  # Remove 'Free' if it exists
storedata_df['Installs_numeric'] = storedata_df['Installs'].str.replace('[+]', '', regex=True).astype(int)
````

````python
# Calculate average sentiment polarity and subjectivity
aggregated_data = review_df.groupby('App').agg(
    Avg_Sentiment_Polarity=('Sentiment_Polarity', 'mean'),
    Avg_Sentiment_Subjectivity=('Sentiment_Subjectivity', 'mean'),
    Positive_Reviews=('Sentiment', lambda x: (x == 'Positive').sum()),
    Neutral_Reviews=('Sentiment', lambda x: (x == 'Neutral').sum()),
    Negative_Reviews=('Sentiment', lambda x: (x == 'Negative').sum())
).reset_index()

# Create the new DataFrame
new_review_df = pd.DataFrame(aggregated_data)
````

````python
df = pd.merge(storedata_df, new_review_df, on='App', how='left')
````

````python
# Group by 'Category' and calculate the mean of each review type
mean_reviews = df.groupby('Category')[['Positive_Reviews', 'Neutral_Reviews', 'Negative_Reviews']].mean().reset_index()

# Sample data
categories = list(mean_reviews['Category'])
positive_reviews = list(mean_reviews['Positive_Reviews'])
neutral_reviews = list(mean_reviews['Neutral_Reviews'])
negative_reviews = list(mean_reviews['Negative_Reviews'])

# Define the number of categories
num_categories = len(categories)

# Create an array of indices for each category
indices = np.arange(num_categories)

# Define the height of each bar
bar_height = 0.75

# Create the stacked bars
plt.figure(figsize=(10, 10))
p1 = plt.barh(indices, positive_reviews, bar_height, label='Positive Reviews', color='#97BE5A')
p2 = plt.barh(indices, neutral_reviews, bar_height, left=positive_reviews, label='Neutral Reviews', color='#FFC96F')
p3 = plt.barh(indices, negative_reviews, bar_height, left=np.array(positive_reviews) + np.array(neutral_reviews), label='Negative Reviews', color='#EE4E4E')

# Add labels, title, and legend
plt.ylabel('Category')
plt.xlabel('Average Number of Reviews')
plt.title('Reviews by Category')
plt.yticks(indices, categories)
plt.legend()
plt.tight_layout()  # Adjust layout to prevent clipping of labels

# Show plot
plt.show()
````
